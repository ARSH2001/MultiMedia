```python
# import matplotlib
# matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import numpy as np
```

```python
def color_change(I, w, new_color):
  plt.imshow(I)
  clicked = plt.ginput(1, show_clicks=True)
  print("clicked", clicked)
  plt.close(1)
  x = int(clicked[0][0])
  y = int(clicked[0][1])
  one = I[y][x][0]
  two = I[y][x][1]
  three = I[y][x][2]
  new_image = np.zeros(I.shape)
  for i in range(I.shape[0]):
    for j in range(I.shape[1]):
      condition = (one-w < I[i][j][0] < one+w) and (two-w < I[i][j][1] < two+w) and \
      (three-w < I[i][j][2] < three+w)
      if condition:
        new_image[i][j] = new_color
      else:
        new_image[i][j] = I[i][j]
  return new_image
```

```python
#%matplotlib tk
Original_image1 = plt.imread("Test08.jpg")
Test1 = color_change(Original_image1, 10, [224, 12, 21]).astype(int)
plt.imshow(Test1)
```
