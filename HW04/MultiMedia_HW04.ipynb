```python
def H4_Shadow(I, s, d):
  #Divide distabce between -d to d into 59 parts
  x = np.linspace(-d, d, 2*s)
  #Create a kernal using normalized cdf of x 
  kernel1d = np.diff(st.norm.cdf(x))
  #Create a kernel using outer product of kernel1 to itself
  kernel2d = np.outer(kernel1d, kernel1d)
  #Create gaussian kernel 
  gaussian_kernel = kernel2d/kernel2d.sum()
  #Create an array for produces one with size 39*39 
  ones = np.zeros((2*s-1,2*s-1))
  #Identify ones usign two for loops , if i & j  in range (s-1,2*s-1) then shadow will be up and left -> this method usign for two lsat images
  for i in range(0,s):
    for j in range(0,s):
      ones[i][j] = 1
  #Create our kernel using procution of gaussian kernel and ones array 
  our_kernel = np.multiply(ones,gaussian_kernel)
  #Create convolution array usign convolce of image I and our kernel 
  conv = sg.convolve2d(I, our_kernel, mode='same')
  #Normalize convolution array
  conv_norm = conv / np.sum(our_kernel)
  #Create our result using sum of image I and normalized convolution
  result = 0.5*(I + conv_norm)
  return result
```

```python
def H4_hybrid(I,J,m,n):
  #Low gaussian kernel with Standard deviation = m
  LGK = ndimage.gaussian_filter(I, m)
  #High gaussian kernel with Standard deviation = n
  HGK = ndimage.gaussian_filter(J, n)
  #High frequency image
  HF = J - HGK
  #Final result
  result = LGK + HF
  #Return results 
  return LGK,HGK,HF,result
```
